Linear Regression With Python


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
X = [1,2,3,4]
Y = [2,4,6,8]
```

```python
plt.scatter(X,Y)
```

```python
alpha = 0.0001
iters = 600

theta =np.array([[3,3,3,3]])
```

```python
def computeCost(X,Y,theta):
    inner = np.power(((X @ theta.T )- Y),2)
    return np.sum(inner)/ (2*len(X))
```

```python
def gradientDescent(X, Y, theta, alpha, iters):
    for i in range(iters):
        cost = computeCost(X, Y, theta)
        theta = theta - (alpha/len(X)) * np.sum((X @ theta.T - Y), axis=0)

        if i % 50 == 0: # just look at cost every ten loops for debugging
            print(cost)
            print(theta)
            print((alpha/len(X)) * np.sum((X @ theta.T - Y), axis=0))

    return (theta, cost)
```

```python
alpha = 0.0001
iters = 650

theta = np.array([[3, 3, 3, 3]])
g, cost = gradientDescent(X, Y, theta, alpha, iters)
print(g, cost)
```

```python
plt.scatter(X, Y)
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = g[0][0] * x_vals #the line equation
plt.plot(x_vals, y_vals, '--')
plt.legend()
```
